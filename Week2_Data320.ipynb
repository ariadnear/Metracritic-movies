### Retrieve all the text from Metacritic website - Year 2015

```python
# Retrieve the text of the metacritic page - Year 2015

import urllib3
import certifi
year = 2015
page = 1

link = f"https://www.metacritic.com/browse/movie/all/all/2013/metascore/?page=1"

# Construct a Python request
http = urllib3.PoolManager(ca_certs=certifi.where())

# Inititate a web request
response = http.request('GET', link, headers={'User-Agent': 'Mozilla/5.0'})

# Convert the result from raw bytes to text
datastring = str(response.data, "utf-8")

# Check whether it succeded and how much data we got from the site
print(f"Fetched {len(datastring)} characters from {link}: {response.status}")
```

### Retrieve Specific Movie Data

```python
# Retrieve the title of movies from the text of the metacritic page using regex.
import re

movie_title = re.findall(r'<div data-title="(.*?)"', datastring)
print(f"Found {len(movie_title)} bullet points in this metacritic page. The first result is {movie_title[0]}")
```

```python
# Retrieve the movie release date.
release_date = re.findall(r'<span class="u-text-uppercase">\s+(.*)\s+<\/span>', datastring)
print(f"Found {len(release_date)} bullet points in this metacritic page. The first result is {release_date[0]}")
```

```python
# Retrieve the description of movies.
movie_description = re.findall(r'<div class="c-finderProductCard_description"><span>(.*?)<', datastring)
print(f"Found {len(movie_description)} bullet points in this metacritic page. The first result is {movie_description[0]}")
```

```python
# Retrieve the metascore of movies.
metascore = re.findall(r'<div title="Metascore (.*?)"', datastring)
print(f"Found {len(metascore)} bullet points in this metacritic page. The first result is {metascore[0]}")
```

```python
# Retrieve the thumbnail photo URL.
# When I did this assignment they used to have the movie poster picture after the metascore. Now they have a metacritic logo.
thumbnail = re.findall(r'<img[^>]+src="([^"]+)"[^>]+>', datastring)
print(f"Found {len(thumbnail)} bullet points in this metacritic page. The first result is {thumbnail[0]}")

# Replace &amp; with &
#fixed_thumbnails = []
#for url in thumbnail:
#    url = url.replace("&amp;", "&")
#    fixed_thumbnails.append(url)

# Print only the first URL after the change. It worked perfect.
#print(fixed_thumbnails[0])
```
